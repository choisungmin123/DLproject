##8. 파생변수 생성 & 변수 확인

기존의 여러 개별 피쳐들을 합쳐서 **총 합계 관련 파생변수(Total 변수)**를 만들어 주었습니다.

예를 들어, 지하실, 1층, 2층 면적을 합쳐 TotalSF를 생성해 주었고, 욕실 개수, 현관 면적 등도 합산하여 집의 전체 규모나 편의 시설을 더 잘 반영하도록 했습니다.

또한, 특정 시설이나 특징의 존재 여부를 0과 1로 나타내는 이진 변수(haspool, hasgarage 등)를 생성했습니다.

예를 들어, 수영장 면적이 0보다 크면 haspool 값이 1(수영장 있음), 0이면 수영장 없음으로 표시하는 식입니다.

###8-2 수치형 변수와 타깃 간 선형성 확인 및 변수 선정
여러 수치형 변수와 타깃 변수(주택가격) 간의 선형 관계를 시각화였습니다.

각 변수별로 산점도와 함께 회귀선을 확인하여 변수와 타깃 간의 관계를 직관적으로 파악했습니다.

또한, 피어슨 상관계수(Pearson correlation coefficient)를 계산하여 변수와 타깃 간의 선형 상관 정도를 수치로 평가했습니다.
예를 들어, li_num[11] 변수와 타깃 간의 상관계수는 약 0.62로, 매우 강한 양의 상관관계를 보였습니다.

이렇게 시각적, 통계적으로 선형성이 강한 변수들만을 선별하여 strong_num 리스트에 저장하였습니다.
이 변수들은 모델에 중요한 피처로 사용될 예정입니다.

###8-3.범주형 변수와 타깃 변수 간 관계 분석

여러 범주형 변수에 대해 박스 플롯(Boxplot)을 그려 각 범주별 주택가격(타깃 변수)의 분포 차이를 시각적으로 확인하였습니다.

이를 통해 타깃 변수에 영향을 많이 미치는 범주형 변수들을 구분할 수 있었습니다.

선별된 주요 범주형 변수들은 strong_obj 리스트에 저장하였으며,

이 변수들에 대해서는 보다 세밀한 분포 확인을 위해 **바이올린 플롯(Violin plot)**을 그려 각 범주 내 주택가격의 분포와 밀도 차이를 시각화하였습니다.

##9. 왜도 처리 및 정규화

데이터의 왜도(skewness)를 확인한 결과, 왜도 절댓값이 0.5 이상인 변수들만 따로 선별하여 skew_feat 리스트에 저장했습니다.

예를 들어 MiscVal, PoolArea, LotArea 등은 매우 높은 왜도를 가지며, 정규분포와 크게 벗어나 있음을 확인했습니다.

이러한 변수들은 머신러닝 및 딥러닝 모델의 성능에 영향을 줄 수 있으므로, Box-Cox 변환을 통해 정규화(normalization) 하였습니다.

정규화는 boxcox1p 함수를 사용하였고, λ(lambda) 값은 0.15로 설정했습니다.

###9-1. 범주형 변수 처리 (One-Hot Encoding)

앞서 Label Encoding을 적용하지 않은 범주형 변수들에 대해는 get_dummies() 함수를 사용하여 더미 변수(one-hot encoding) 를 생성했습니다.

이는 범주형 변수를 모델이 인식 가능한 수치 형태로 바꾸기 위한 전처리 작업입니다.

이 과정을 통해 최종적으로 전체 피처 수는 230개로 확장되었으며, 모델 학습에 필요한 모든 전처리가 완료되었습니다.
